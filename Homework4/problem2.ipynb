```python
import matplotlib.pyplot as plt

with open('problem2.dat', 'r') as file:
    data = [list(map(float, line.split())) for line in file]

x_values = [row[0] for row in data]
y1_values = [row[1] for row in data]
y2_values = [row[2] for row in data]

plt.plot(x_values, y1_values, label='Reduction Time')
plt.plot(x_values, y2_values, label='Total Time')

plt.xlabel('Processes')
plt.ylabel('Time')
plt.title('Reduction and Total Time Vs. Number of Processes')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('problem2.dat')

cores = data[:, 0]
reduction_time = data[:, 1]
total_time = data[:, 2]

plt.figure(figsize=(10, 6))
plt.plot(cores, total_time, marker='o', label='Total Walltime', color='blue')

plt.plot(cores, reduction_time, marker='o', label='Reduction Walltime', color='orange')

plt.xlabel('Cores')
plt.ylabel('Walltime (seconds)')
plt.title('Cores vs Total and Reduction Walltime')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('problem2.dat')

cores = data[:, 0]
reduction_time = data[:, 1]
total_time = data[:, 2]

total_slope, total_intercept = np.polyfit(cores, total_time, 1)
reduction_slope, reduction_intercept = np.polyfit(cores, reduction_time, 1)

plt.figure(figsize=(10, 6))
plt.plot(cores, total_time, marker='o', label='Total Walltime', color='blue')
plt.plot(cores, total_slope * cores + total_intercept, linestyle='--', color='blue', label=f'Total Slope: {total_slope:.6f}')

plt.plot(cores, reduction_time, marker='o', label='Reduction Walltime', color='orange')
plt.plot(cores, reduction_slope * cores + reduction_intercept, linestyle='--', color='orange', label=f'Reduction Slope: {reduction_slope:.6f}')

plt.xlabel('Cores')
plt.ylabel('Walltime (seconds)')
plt.title('Cores vs Total and Reduction Walltime')
plt.legend()
plt.grid(True)
plt.show()

print(f'Total Slope: {total_slope:.6f}')
print(f'Reduction Slope: {reduction_slope:.6f}')
```
