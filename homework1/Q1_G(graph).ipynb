```python
# Log-log graph of error vs N

import math
import matplotlib.pyplot as plt
import numpy as np

dat = np.loadtxt('Q1_G.dat', comments='Iteration')


N = dat[:, 0]
error = dat[:, -1]

plt.figure(figsize=(12, 6))
plt.loglog(N, error, '--bo')
plt.grid(True)
plt.xlabel('log N')
plt.ylabel('log error')
```
