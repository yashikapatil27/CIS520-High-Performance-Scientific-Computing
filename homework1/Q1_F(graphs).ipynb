```python
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import math

data_file = '../homework1/data_files'

mean_errors = {}
mean_error_values = []

for N in range(1, 31):
    filename = f'numbers{N}.dat' 
    filepath = os.path.join(data_file, filename)
    with open(filepath, 'r') as file:
        for line in file:
            current_N, error = map(float, line.split())
            if current_N not in mean_errors:
                mean_errors[current_N] = []
            mean_errors[current_N].append(error)

for N, errors in sorted(mean_errors.items()):
    mean_error = sum(errors)/len(errors)
    mean_error_values.append(mean_error)
    print(f'N: {N} \n Mean error: {mean_error} ')
```
